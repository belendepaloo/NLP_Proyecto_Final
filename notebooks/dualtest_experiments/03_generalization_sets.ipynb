```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[1] if Path.cwd().name == "dualtest_experiments" else Path.cwd()
sys.path.append(str(PROJECT_ROOT / "DUALTEST"))

print(PROJECT_ROOT)
```

```python
import pandas as pd

from generalization_sets import build_generalization_sets
from prefixing import split_by_tokens
from target_model import HFLocalTarget
from reference_model import ReferenceModel
from metrics import rlb_score, esb_score
from scoring import score_texts 
```

```python
target = HFLocalTarget(
    model_name="Qwen/Qwen2.5-1.5B",
    device="cuda"
)

reference = ReferenceModel(
    model_name="Qwen/Qwen2.5-0.5B",
    device="cuda"
)
```

```python
gen_sets = build_generalization_sets(
    n_a=5,
    n_b=2,
    domains=("book",)
)

gen_sets["book"]["A"][:2], gen_sets["book"]["B"][:1]
```

```python
df_adv_a = score_texts(
    gen_sets["book"]["A"],
    target=target,
    reference=reference,
    dataset_name="generalization_book_A",
)

df_adv_b = score_texts(
    gen_sets["book"]["B"],
    target=target,
    reference=reference,
    dataset_name="generalization_book_B",
)

df_adv = pd.concat([df_adv_a, df_adv_b], ignore_index=True)
df_adv
```

```python
gen_sets = build_generalization_sets(
    n_a=100,
    n_b=10,
    domains=("book",)
)

df_adv_a = score_texts(
    texts=gen_sets["book"]["A"],
    target=target,
    reference=reference,
    dataset_name="generalization_book_A",
    label=0,
)

df_adv_a["set"] = "A"
df_adv_a["domain"] = "book"

df_adv_b = score_texts(
    texts=gen_sets["book"]["B"],
    target=target,
    reference=reference,
    dataset_name="generalization_book_B",
    label=0,
)

df_adv_b["set"] = "B"
df_adv_b["domain"] = "book"

df_adv = pd.concat([df_adv_a, df_adv_b], ignore_index=True)

df_adv
```

```python
out_dir = PROJECT_ROOT / "results"
out_dir.mkdir(exist_ok=True)

out_path = out_dir / "generalization_book_qwen15_qwen05.csv"
df_adv.to_csv(out_path, index=False)

print(out_path)
```
